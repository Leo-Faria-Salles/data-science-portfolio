```python
import pandas as pd

dados = pd.read_csv('Advertising.csv', header=(0))
print(dados)
```

```python
#Vamos exluir a primeira coluna
data = dados.drop(dados.columns[[0]], axis=1)
print(data)
```

```python
#Obtendo amatriz de correlação dos dados
corr = data.corr()
print(corr)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7, 7))
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns)
plt.suptitle('Correlação entre as variáveis', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show
```
